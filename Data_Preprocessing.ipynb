## **8.2. Imbalanced Data**

Karena kedua kelas Disease dan Non-Disease belum seimbang, maka akan digunakan teknik oversampling untuk menyeimbangkan data. Tujuannya untuk meningkatkan keseimbangan kelas dalam dataset, sehingga model pembelajaran mesin memiliki performa yang lebih baik dalam mengklasifikasikan kelas minoritas. Dengan jumlah sampel yang lebih seimbang antara kelas "Disease" dan "Non-Disease", model akan lebih mungkin untuk belajar dengan baik dari kedua kelas dan menghasilkan prediksi yang lebih akurat.


```python
from imblearn.over_sampling import SMOTE

# Memisahkan fitur dan target
X = data_heart.drop('target', axis=1)
y = data_heart['target']

# Melakukan oversampling dengan SMOTE
oversampler = SMOTE()
X_resampled, y_resampled = oversampler.fit_resample(X, y)

data_heart = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.DataFrame(y_resampled, columns=['target'])], axis=1)
```

```python
# Periksa ukuran setelah oversampling
proporsi_heart_dis = data_heart.groupby('target', as_index = False).agg(total = ('target', 'count')).replace({0: 'No Disease', 1: 'Disease'})

proporsi_heart_dis
```

## **8.4. Handling Outlier**

Karena ditemukan beberapa outlier pada kolom trestbps, chol, thalach, dan oldpeak maka akan diatasi dengan mengganti semua nilai outlier ke nilai batas atas atau batas bawah.

```python
# Import library operasi numerik
import numpy as np
```

```python
# Proses penentuan ambang batas outlier
def imputasi_outlier(data, kolom):
    Q1 = data[kolom].quantile(q = 0.25)
    Q3 = data[kolom].quantile(q = 0.75)
    IQR = Q3 - Q1
    lower_fence = Q1 - 1.5 * IQR
    upper_fence = Q3 + 1.5 * IQR

    conditions = [data[kolom] < lower_fence, data[kolom] > upper_fence]
    choices = [lower_fence, upper_fence]

    data[kolom] = np.select(conditions, choices, default = data[kolom])
    return(data)
```

Diketahui bahwa terdapat outlier pada kolom selain 'age'.

```python
for column in ['trestbps', 'chol', 'thalach', 'oldpeak']:
    data_heart = imputasi_outlier(data = data_heart, kolom = column)
    data_heart.shape
```

# **10. Deep Dive Analysis**

Carilah fitur apa yang mungkin mempengaruhi seseorang terkena heart disease dan interpretasikan.

**1. age_category**

Fitur ini memungkinkan untuk memudahkan memahami hubungan antara usia dan risiko penyakit jantung. Dengan mengelompokkan usia ke dalam beberapa kategori-kategori, pengguna data dapat dengan cepat mengidentifikasi pola atau tren yang terkait dengan usia dalam konteks risiko penyakit jantung.
