## <b>No 2, 3, 4</b>

Hitunglah BMI setiap siswa dengan rumus BMI = Berat badan (kg)/(Tinggi Badan (m))^2

Buatlah program untuk menghitung BMI!

Update dataframe df dengan menambahkan hasil perhitungan soal nomor 2, beri nama kolom BMI.


```python
# Hitung BMI untuk setiap siswa dan menambahkan kolom baru 'BMI'
df_6['BMI'] = df_6['Berat Badan (kg)'] / ((df_6['Tinggi Badan (cm)'] / 100) ** 2)

# Tampilkan data
display(df_6)
```

## <b>No 5</b>

Buatlah status BMI untuk setiap siswa dari kategori berikut, lalu simpan data statusnya dengan nama kolom Status BMI.
- Normal: 18.5-24.9
- Overweight: 25.0-29.9
- Obesity: 30.0-34.9
- Underweight: < 18.5


```python
# Membuat fungsi kategori BMI
def kategori_bmi(bmi):
  if 18.5 <= bmi < 24.9:
      return "Normal"
  elif 25.0 <= bmi < 29.9:
      return "Overweight"
  elif 30.0 <= bmi < 34.9:
      return "Obesity"
  elif bmi < 18.5:
    return "Underweight"
  else :
    # lebih dari 34.9
    return "Very Obesity"

# Tambahkan kolom dengan kategori BMI
df_6['Status BMI'] = df_6['BMI'].apply(kategori_bmi)

# Tampilkan data
display(df_6)
```

Tabel di atas adalah semua data yang ada pada dataframe, setelah ditambahkan kolom Status BMI dari pengkategorian pada fungsi.

# <b>Soal 2</b>

Perusahaan Investasi Jamin-Sugih Co kami memberikan penawaran investasi terbaik dengan pembagian keuntungan tetap sebesar 25% setiap tahun.

Buatlah program Python yang dapat menerima input dari pengguna dalam bentuk:

Modal Investasi (bilangan bulat antara 0-1.000.000.000)

Jangka Waktu Investasi (bilangan bulat antara 1-25 dalam tahun)

Selain itu, program akan menampilkan error message bahwa modal atau tenure tidak sesuai ketentuan.

Program akan menampilkan informasi akumulasi investasi setiap tahun, menggunakan iterasi untuk setiap tahun.

```python
# Minta pengguna untuk memasukkan modal investasi dan jangka waktu investasi
modal = int(input("Masukkan Modal Investasi (0-1.000.000.000): "))
tenure = int(input("Masukkan Jangka Waktu Investasi (1-25 tahun): "))

# Periksa apakah modal investasi dan jangka waktu investasi berada dalam rentang yang ditentukan
if not(0 <= modal <= 1000000000) or not(1 <= tenure <= 25):
    print("Modal atau tenure tidak sesuai ketentuan.")
else:
    # Jika modal investasi dan jangka waktu investasi valid, gunakan loop untuk menghitung dan mencetak total investasi untuk setiap tahun
    for i in range(1, tenure + 1):
        modal *= 1.25
        print("Tahun {}, total investasi : {}".format(i, modal))
```
